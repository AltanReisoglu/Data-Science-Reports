# 🎬 LongCat-Video — Uzun Video Üretimi

**Model:** [meituan-longcat/LongCat-Video](https://huggingface.co/meituan-longcat/LongCat-Video)  
**Çözünürlük:** 720P  
**Özellik:** Uzun süreli tutarlı video üretimi  
**İndirme:** 732  
**Like:** 450  

Meituan tarafından geliştirilen LongCat-Video, uzun süreli ve tutarlı video üretimi konusunda uzmanlaşmıştır.

```python
# !pip install torch torchvision diffusers transformers accelerate imageio[ffmpeg]
```

```python
import torch
from diffusers import DiffusionPipeline
from diffusers.utils import export_to_video

MODEL_ID = "meituan-longcat/LongCat-Video"

# Not: Model custom pipeline kullanabilir, model kartını kontrol edin
pipe = DiffusionPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16,
    trust_remote_code=True,
)
pipe.enable_model_cpu_offload()
```

```python
prompt = "A long continuous shot following a river through a valley, starting from mountain peaks and flowing down to a village, autumn colors, aerial drone footage style"

output = pipe(
    prompt=prompt,
    num_frames=161,       # Uzun video
    height=480,
    width=720,
    num_inference_steps=50,
    guidance_scale=7.0,
    generator=torch.Generator(device="cpu").manual_seed(42),
)

export_to_video(output.frames[0], "longcat_output.mp4", fps=24)
print("Video kaydedildi: longcat_output.mp4")
```
