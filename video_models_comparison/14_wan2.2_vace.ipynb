# 🎬 Wan2.2-VACE-Fun — Kontrollü Video Üretimi

**Model:** [alibaba-pai/Wan2.2-VACE-Fun-A14B](https://huggingface.co/alibaba-pai/Wan2.2-VACE-Fun-A14B)  
**Parametre:** 14B  
**Çözünürlük:** 720P  
**Özellik:** VACE (Video and Controllable Enhancement) — T2V + kontrol mekanizmaları  
**İndirme:** 1.65k  
**Like:** 108  

Alibaba PAI takımının VACE framework'ü ile Wan2.2 üzerine kurulmuş kontrollü video üretim modeli. Edge maps, depth maps gibi koşullarla yönlendirilmiş video üretimi sağlar.

```python
# !pip install torch torchvision diffusers transformers accelerate imageio[ffmpeg] pillow
```

```python
import torch
from diffusers import DiffusionPipeline
from diffusers.utils import export_to_video

MODEL_ID = "alibaba-pai/Wan2.2-VACE-Fun-A14B"

pipe = DiffusionPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16,
    trust_remote_code=True,
)
pipe.enable_model_cpu_offload()
```

```python
prompt = "A woman dances gracefully in a flowing red dress, professional studio lighting, slow motion, high fashion, editorial style photography"
negative_prompt = "blurry, low quality, distorted, watermark"

# Not: VACE modunda ek kontrol sinyalleri (edge/depth map) gönderilebilir
# Temel T2V modu için:
output = pipe(
    prompt=prompt,
    negative_prompt=negative_prompt,
    num_frames=81,
    height=480,
    width=720,
    num_inference_steps=50,
    guidance_scale=5.0,
    generator=torch.Generator(device="cpu").manual_seed(42),
)

export_to_video(output.frames[0], "wan_vace_output.mp4", fps=24)
print("Video kaydedildi: wan_vace_output.mp4")
```
