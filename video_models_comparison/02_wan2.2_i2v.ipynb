# 🖼️➡🎬 Wan2.2-I2V-A14B — Image-to-Video

**Model:** [Wan-AI/Wan2.2-I2V-A14B](https://huggingface.co/Wan-AI/Wan2.2-I2V-A14B)  
**Parametre:** 14B (MoE)  
**Çözünürlük:** 720P @ 24fps  
**VRAM:** ~24GB+  

Wan2.2'nin Image-to-Video versiyonu. Bir görsel verip, o görseli hareketlendiren yüksek kaliteli videolar üretir.

```python
# !pip install torch torchvision diffusers transformers accelerate sentencepiece imageio[ffmpeg] pillow
```

```python
import torch
from diffusers import WanImageToVideoPipeline
from diffusers.utils import export_to_video, load_image

MODEL_ID = "Wan-AI/Wan2.2-I2V-A14B-Diffusers"

pipe = WanImageToVideoPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16,
)
pipe.enable_model_cpu_offload()
```

```python
# Girdi görseli — Kendi görselini buraya koy
image = load_image("https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/diffusers/astronaut.png")
image = image.resize((720, 480))
image
```

```python
prompt = "The astronaut slowly turns their head, the visor reflecting starlight, subtle floating motion in zero gravity, cinematic lighting"
negative_prompt = "blurry, low quality, distorted"

output = pipe(
    image=image,
    prompt=prompt,
    negative_prompt=negative_prompt,
    num_frames=81,
    height=480,
    width=720,
    guidance_scale=5.0,
    num_inference_steps=50,
    generator=torch.Generator(device="cpu").manual_seed(42),
)

export_to_video(output.frames[0], "wan2.2_i2v_output.mp4", fps=24)
print("Video kaydedildi: wan2.2_i2v_output.mp4")
```
