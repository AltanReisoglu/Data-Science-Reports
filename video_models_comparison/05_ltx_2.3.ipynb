# 🎬 LTX-2.3 — Image-to-Video (Audio+Video)

**Model:** [Lightricks/LTX-2.3](https://huggingface.co/Lightricks/LTX-2.3)  
**Parametre:** ~21B  
**Çözünürlük:** 4K @ 50fps  
**Özellik:** Ses ve video eşzamanlı üretim  
**Lisans:** Apache 2.0  
**VRAM:** ~40GB+ (fp16), fp8/nvfp4 versiyonları mevcut  

LTX-2.3, Lightricks'in en yeni modeli. Ses ve video eşzamanlı üretebilen ilk açık kaynak modellerden biri. 4K çözünürlükte 50fps destekler. 1.13M indirme ile trending #1.

```python
# !pip install torch torchvision diffusers transformers accelerate imageio[ffmpeg] pillow
```

```python
import torch
from diffusers import LTXPipeline
from diffusers.utils import export_to_video, load_image

# LTX-2.3 fp8 versiyonu (daha düşük VRAM)
MODEL_ID = "Lightricks/LTX-2.3-fp8"

pipe = LTXPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.bfloat16,
)
pipe.enable_model_cpu_offload()
```

```python
# Girdi görseli
image = load_image("https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/diffusers/astronaut.png")
image = image.resize((768, 512))
image
```

```python
prompt = "The astronaut floats gracefully through a space station corridor, reaching out to grab a floating tool, dynamic camera follows the motion"

output = pipe(
    prompt=prompt,
    # image=image,  # I2V modu için
    num_frames=97,
    height=512,
    width=768,
    num_inference_steps=50,
    guidance_scale=7.5,
    generator=torch.Generator(device="cpu").manual_seed(42),
)

export_to_video(output.frames[0], "ltx_2.3_output.mp4", fps=24)
print("Video kaydedildi: ltx_2.3_output.mp4")
```
