# 🎬 HunyuanVideo 1.5 — Text/Image-to-Video

**Model:** [tencent/HunyuanVideo-1.5](https://huggingface.co/tencent/HunyuanVideo-1.5)  
**Parametre:** 8.3B (daha hafif)  
**Çözünürlük:** 480P-720P + Super-Resolution → 1080P  
**VRAM:** ~13.6GB (RTX 4090, offload ile)  

HunyuanVideo 1.5, önceki versiyonun daha hafif ama güçlü versiyonu. Consumer-grade GPU'larda çalışabilir. Entegre video super-resolution DiT ile 1080P'ye upscale yapabilir.

```python
# !pip install torch torchvision diffusers transformers accelerate imageio[ffmpeg]
```

```python
import torch
from diffusers import HunyuanVideoPipeline, HunyuanVideoTransformer3DModel
from diffusers.utils import export_to_video

MODEL_ID = "tencent/HunyuanVideo-1.5"

transformer = HunyuanVideoTransformer3DModel.from_pretrained(
    MODEL_ID,
    subfolder="transformer",
    torch_dtype=torch.bfloat16,
)

pipe = HunyuanVideoPipeline.from_pretrained(
    MODEL_ID,
    transformer=transformer,
    torch_dtype=torch.float16,
)
pipe.vae.enable_tiling()
pipe.enable_model_cpu_offload()
```

```python
prompt = "A futuristic city at night with neon lights reflecting on wet streets, flying cars passing between skyscrapers, cyberpunk aesthetic, cinematic"

output = pipe(
    prompt=prompt,
    num_frames=61,
    height=480,
    width=848,
    num_inference_steps=50,
    guidance_scale=6.0,
    generator=torch.Generator(device="cpu").manual_seed(42),
)

export_to_video(output.frames[0], "hunyuan_1.5_output.mp4", fps=15)
print("Video kaydedildi: hunyuan_1.5_output.mp4")
```
