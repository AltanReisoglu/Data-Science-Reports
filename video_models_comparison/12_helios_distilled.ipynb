# 🎬 Helios-Distilled — Hızlı Text-to-Video

**Model:** [BestWishYsh/Helios-Distilled](https://huggingface.co/BestWishYsh/Helios-Distilled)  
**İndirme:** 8.95k  
**Like:** 42  

Helios-Distilled, distillation ile hızlandırılmış bir text-to-video modeli. Daha az inference step ile kaliteli video üretir.

```python
# !pip install torch torchvision diffusers transformers accelerate imageio[ffmpeg]
```

```python
import torch
from diffusers import DiffusionPipeline
from diffusers.utils import export_to_video

MODEL_ID = "BestWishYsh/Helios-Distilled"

pipe = DiffusionPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16,
    trust_remote_code=True,
)
pipe.enable_model_cpu_offload()
```

```python
prompt = "A majestic eagle soaring above snow-capped mountains, wind ruffling its feathers, panoramic landscape below, nature documentary quality"

output = pipe(
    prompt=prompt,
    num_frames=81,
    height=480,
    width=720,
    num_inference_steps=25,  # Distilled = daha az step
    guidance_scale=5.0,
    generator=torch.Generator(device="cpu").manual_seed(42),
)

export_to_video(output.frames[0], "helios_distilled_output.mp4", fps=24)
print("Video kaydedildi: helios_distilled_output.mp4")
```
