# 🎬 LTX-2 — Text/Image-to-Video

**Model:** [Lightricks/LTX-2](https://huggingface.co/Lightricks/LTX-2)  
**Çözünürlük:** 4K @ 50fps  
**Özellik:** Ses+video birleşik üretim, multi-keyframe, IC-LoRA, 20 sn'ye kadar continuous video  
**Lisans:** Apache 2.0  
**İndirme:** 1.17M  

Production-ready açık kaynak video üretim modeli. Distilled hybrid mimari ile yüksek kalite ve hız.

```python
# !pip install torch torchvision diffusers transformers accelerate imageio[ffmpeg]
```

```python
import torch
from diffusers import LTXPipeline
from diffusers.utils import export_to_video

MODEL_ID = "Lightricks/LTX-2"

pipe = LTXPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.bfloat16,
)
pipe.enable_model_cpu_offload()
```

```python
prompt = "A time-lapse of a flower blooming from bud to full bloom, soft natural light, macro photography, shallow depth of field, nature documentary style"

output = pipe(
    prompt=prompt,
    num_frames=97,
    height=512,
    width=768,
    num_inference_steps=50,
    guidance_scale=7.5,
    generator=torch.Generator(device="cpu").manual_seed(42),
)

export_to_video(output.frames[0], "ltx_2_output.mp4", fps=24)
print("Video kaydedildi: ltx_2_output.mp4")
```
