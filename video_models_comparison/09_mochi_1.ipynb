# 🎬 Mochi 1 — Text-to-Video

**Model:** [genmo/mochi-1-preview](https://huggingface.co/genmo/mochi-1-preview)  
**Parametre:** 10B  
**Çözünürlük:** 848x480  
**Mimari:** Asymmetric Diffusion Transformer (AsymmDiT)  
**VRAM:** ~40GB (fp16)  

Genmo tarafından geliştirilen Mochi 1, prompt'a sadık kalma ve hareket kalitesi konusunda öne çıkar. LoRA fine-tuning desteği sunar.

```python
# !pip install torch torchvision diffusers transformers accelerate imageio[ffmpeg]
```

```python
import torch
from diffusers import MochiPipeline
from diffusers.utils import export_to_video

MODEL_ID = "genmo/mochi-1-preview"

pipe = MochiPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.bfloat16,
)
pipe.enable_model_cpu_offload()
pipe.enable_vae_tiling()
```

```python
prompt = "A hand reaches out to pick up a glowing crystal from a cave floor, magical particles swirl around it, dramatic lighting, fantasy cinematic"

output = pipe(
    prompt=prompt,
    num_frames=84,         # 6 sn @30fps compatible (mochi her 6 frame atlama yapar)
    height=480,
    width=848,
    num_inference_steps=64,
    guidance_scale=4.5,
    generator=torch.Generator(device="cpu").manual_seed(42),
)

export_to_video(output.frames[0], "mochi_1_output.mp4", fps=30)
print("Video kaydedildi: mochi_1_output.mp4")
```
