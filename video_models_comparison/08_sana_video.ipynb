# 🎬 SANA-Video 2B — Text-to-Video

**Model:** [Efficient-Large-Model/SANA-Video_2B_720p](https://huggingface.co/Efficient-Large-Model/SANA-Video_2B_720p)  
**Parametre:** 2B (çok hafif!)  
**Çözünürlük:** 720P  
**VRAM:** ~8-12GB  

SANA-Video, sadece 2B parametre ile yüksek kaliteli 720P video üretebilen son derece verimli bir model. Consumer GPU'larda rahatça çalışır. Mart 2026'da trending.

```python
# !pip install torch torchvision diffusers transformers accelerate imageio[ffmpeg]
```

```python
import torch
from diffusers import DiffusionPipeline
from diffusers.utils import export_to_video

MODEL_ID = "Efficient-Large-Model/SANA-Video_2B_720p"

pipe = DiffusionPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16,
    trust_remote_code=True,
)
pipe.enable_model_cpu_offload()
```

```python
prompt = "A drone shot flying over a tropical rainforest, mist rising from the canopy, a waterfall visible in the distance, golden morning light, National Geographic style"

output = pipe(
    prompt=prompt,
    num_frames=81,
    height=480,
    width=720,
    num_inference_steps=50,
    guidance_scale=7.0,
    generator=torch.Generator(device="cpu").manual_seed(42),
)

export_to_video(output.frames[0], "sana_video_output.mp4", fps=24)
print("Video kaydedildi: sana_video_output.mp4")
```
