# 🎬 Wan2.2-T2V-A14B — Text-to-Video

**Model:** [Wan-AI/Wan2.2-T2V-A14B](https://huggingface.co/Wan-AI/Wan2.2-T2V-A14B)  
**Parametre:** 14B (Mixture-of-Experts)  
**Çözünürlük:** 720P / 1080P @ 24fps  
**Mimari:** DiT + MoE  
**VRAM:** ~24GB+ (fp16), quantized GGUF versiyonları daha düşük  

Alibaba Tongyi Lab tarafından geliştirilen Wan2.2, Mixture-of-Experts mimarisi sayesinde yüksek kaliteli sinematik videolar üretir. Aydınlatma, kompozisyon ve renk tonu için detaylı etiketlerle eğitilmiştir.

```python
# !pip install torch torchvision diffusers transformers accelerate sentencepiece imageio[ffmpeg]
```

```python
import torch
from diffusers import WanPipeline
from diffusers.utils import export_to_video

MODEL_ID = "Wan-AI/Wan2.2-T2V-A14B-Diffusers"

pipe = WanPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16,
)
pipe.enable_model_cpu_offload()  # VRAM tasarrufu
```

```python
prompt = "A cinematic aerial shot of a coastal city at golden hour, with waves crashing against the shore, warm sunlight reflecting off glass buildings, slow camera pan, 4K quality, photorealistic"
negative_prompt = "blurry, low quality, distorted, watermark, text"

output = pipe(
    prompt=prompt,
    negative_prompt=negative_prompt,
    num_frames=81,        # ~3.4 saniye @24fps
    height=480,
    width=720,
    guidance_scale=5.0,
    num_inference_steps=50,
    generator=torch.Generator(device="cpu").manual_seed(42),
)

export_to_video(output.frames[0], "wan2.2_t2v_output.mp4", fps=24)
print("Video kaydedildi: wan2.2_t2v_output.mp4")
```
