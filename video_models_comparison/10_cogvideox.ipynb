# 🎬 CogVideoX-5B — Text-to-Video

**Model:** [THUDM/CogVideoX-5b](https://huggingface.co/THUDM/CogVideoX-5b)  
**Parametre:** 5B  
**Çözünürlük:** 720x480 @ 8fps  
**Uzunluk:** ~6 saniye  
**VRAM:** ~18GB (bf16)  

Tsinghua Üniversitesi (THUDM) tarafından geliştirilen CogVideoX, Expert Transformer mimarisi ile text-to-video üretiminde güçlü sonuçlar sunar.

```python
# !pip install torch torchvision diffusers transformers accelerate imageio[ffmpeg]
```

```python
import torch
from diffusers import CogVideoXPipeline
from diffusers.utils import export_to_video

MODEL_ID = "THUDM/CogVideoX-5b"

pipe = CogVideoXPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.bfloat16,
)
pipe.enable_model_cpu_offload()
pipe.vae.enable_tiling()
```

```python
prompt = "A close-up shot of a butterfly landing on a flower, detailed wing patterns visible, soft bokeh background, macro photography, natural light"

output = pipe(
    prompt=prompt,
    num_frames=49,
    height=480,
    width=720,
    num_inference_steps=50,
    guidance_scale=6.0,
    generator=torch.Generator(device="cpu").manual_seed(42),
)

export_to_video(output.frames[0], "cogvideox_output.mp4", fps=8)
print("Video kaydedildi: cogvideox_output.mp4")
```
