# 🎬 SkyReels-V3 — Image/Audio-to-Video

**Model:** [Skywork/SkyReels-V3-A2V-19B](https://huggingface.co/Skywork/SkyReels-V3-A2V-19B)  
**Parametre:** 19B  
**Çözünürlük:** 720P  
**Özellik:** Sonsuz uzunlukta video üretimi (Diffusion Forcing), kamera kontrolü, lip-sync, ses efektleri  
**VRAM:** ~40GB+  

Skywork AI'ın SkyReels-V3 modeli, 'Diffusion Forcing' framework'ü ile sonsuz uzunlukta sinematik videolar üretebilir. Ocak 2026'da yayınlanmıştır.

```python
# !pip install torch torchvision diffusers transformers accelerate imageio[ffmpeg] pillow
```

```python
import torch
from diffusers import DiffusionPipeline
from diffusers.utils import export_to_video, load_image

MODEL_ID = "Skywork/SkyReels-V3-A2V-19B"

# Not: Bu model custom pipeline kullanabilir.
# HuggingFace model kartını kontrol edin.
pipe = DiffusionPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16,
    trust_remote_code=True,
)
pipe.enable_model_cpu_offload()
```

```python
# Girdi görseli
image = load_image("https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/diffusers/astronaut.png")
image = image.resize((720, 480))
image
```

```python
prompt = "The astronaut walks forward on the moon surface, dust particles floating in slow motion, Earth visible in the background, cinematic wide angle shot"

# Not: API, modelin resmi reposundaki kullanım kılavuzuna göre değişebilir
output = pipe(
    image=image,
    prompt=prompt,
    num_frames=81,
    height=480,
    width=720,
    num_inference_steps=50,
    guidance_scale=7.0,
    generator=torch.Generator(device="cpu").manual_seed(42),
)

export_to_video(output.frames[0], "skyreels_v3_output.mp4", fps=24)
print("Video kaydedildi: skyreels_v3_output.mp4")
```
